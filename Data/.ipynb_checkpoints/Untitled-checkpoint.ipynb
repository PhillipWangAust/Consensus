```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import pickle
```

```python
N=100
n_iterations=30
n_probs=30
Path = np.zeros((n_iterations,n_probs))
Cluster = np.zeros((n_iterations,n_probs))
avg_degree = np.zeros((n_iterations,n_probs))
for i in range(0,n_iterations):
    print(i)
    k=0
    for p in np.linspace(0.1,1,n_probs):
        ER = nx.gnp_random_graph(N, p, seed=None)
        while(not nx.is_connected(ER)):
            ER = nx.gnp_random_graph(N, p, seed=None)
        Path[i,k] = nx.average_shortest_path_length(ER)
        Cluster[i,k] = nx.average_clustering(ER)
        avg_degree[i,k] = np.mean(np.array(ER.degree)[:,1])
        k=k+1
        
pickle.dump(Path, open("ER_Path","wb"))
pickle.dump(Cluster,open("ER_cluster","wb"))
```

```python
x = np.linspace(0.1,1,n_probs)
plt.plot(x,np.mean(avg_degree,axis=0))
```

```python
plt.plot(x,np.mean(Cluster,axis=0))
```

```python
N=100
n_iterations=30
n_probs=30
Path = np.zeros((n_iterations,n_probs))
Cluster = np.zeros((n_iterations,n_probs))
avg_degree = np.zeros((n_iterations,n_probs))
for i in range(0,n_iterations):
    print(i)
    k=0
    for p in np.linspace(0,1,n_probs):
        GWS = nx.watts_strogatz_graph(N, 3,p)
        while(not nx.is_connected(GWS)):
            GWS = nx.watts_strogatz_graph(N, 3,p)
        Path[i,k] = nx.average_shortest_path_length(GWS)
        Cluster[i,k] = nx.average_clustering(GWS)
        avg_degree[i,k] = np.mean(np.array(GWS.degree)[:,1])
        k=k+1
        
pickle.dump(Path, open("WS_Path","wb"))
pickle.dump(Cluster,open("WS_cluster","wb"))
```

```python
N=100
n_iterations=30
n_probs=30
Path = np.zeros((n_iterations,n_probs))
Cluster = np.zeros((n_iterations,n_probs))
for i in range(0,n_iterations):
    print(i)
    k=0
    for p in np.linspace(0,1,n_probs):
        GBA = nx.barabasi_albert_graph(N, 1)
        while(not nx.is_connected(GBA)):
            GBA = nx.barabasi_albert_graph(N, 1)
        Path[i,k] = nx.average_shortest_path_length(GBA)
        Cluster[i,k] = nx.average_clustering(GBA)
        k=k+1
        
pickle.dump(Path, open("BA_Path","wb"))
pickle.dump(Cluster,open("BA_cluster","wb"))
```

```python
x = np.linspace(0,1,n_probs)
plt.plot(x,np.mean(Path,axis=0))
```

```python
plt.plot(x,np.mean(Cluster,axis=0))
```

```python
Path = pickle.load(open("WS_Path","rb"))

x = np.linspace(0,1,n_probs)
plt.plot(x,np.mean(Path,axis=0))
plt.grid()
plt.title("Average shortest path length for the Watts-Strogatz model \n"
         +"with re-wire probability p, average of 30 realisations")
plt.xlabel("p: for Watts-Strogatz model (p,N=100,k=3)")
plt.ylabel("Average shortest path length")
plt.savefig("WS shortest path.png",bbox_inches='tight')
```

```python
Path = pickle.load(open("ER_Path","rb"))

x = np.linspace(0,1,n_probs)
plt.plot(x,np.mean(Path,axis=0))
plt.grid()
plt.title("Average shortest path length for the Erdos-Renyi model \n"
         +"with probability p and N=100, average of 30 realisations")
plt.xlabel("p: for Erdos-Renyi model (p,N=100)")
plt.ylabel("Average shortest path length")
plt.savefig("ER shortest path.png",bbox_inches='tight')
```
